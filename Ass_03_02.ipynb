Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a set of points plotted on a horizontal and vertical axes. Scatter plots are important in statistics because they can show the extent of correlation, if any, between the values of observed quantities or phenomena (called variables).

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Chart-1')


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Chart-2')
plt.show()

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Chart-3')


plt.subplot(2,2,4)
plt.plot(x,y4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Chart-4')
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

The bar graph helps to compare the different sets of data among different groups easily. It shows the relationship using two axes, in which the categories are on one axis and the discrete values are on the other axis. 

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])
plt.show()
```
